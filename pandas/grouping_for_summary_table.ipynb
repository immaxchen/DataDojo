```python
import pandas as pd
```

```python
df = pd.DataFrame({'col_1':list('aabbbbbb'),
                   'col_2':list('aabbabab'),
                   'col_3':range(8),
                   'col_4':range(0,16,2)})
df
```

### basic summary

```python
df.groupby(['col_1','col_2'])['col_3','col_4'].mean().rename(columns={'col_3':'mean_3', 'col_4':'mean_4'})
```

### multiple aggregations

```python
df.groupby(['col_1','col_2']).agg( {'col_3':['min','max'], 'col_4':['mean','std']} )
```

### built-in describe

```python
df.groupby(['col_1','col_2'])['col_3','col_4'].describe()
```
