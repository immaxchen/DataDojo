```python
import pandas as pd
```

```python
df = pd.DataFrame({'col_1':list('aabbbbbb'),
                   'col_2':list('aabbabab'),
                   'col_3':range(8),
                   'col_4':range(0,16,2)})
df
```

### get row-number by original order

```python
df['rownum'] = df.groupby(['col_1','col_2']).cumcount()
df
```

### get row-number by specific order

```python
df['rownum'] = df.groupby(['col_1','col_2'])['col_3'].rank(ascending=False).astype(int)
df
```

### get row-number by specific order using sort

```python
df = df.sort_values(['col_3','col_4'], ascending=[False,True])
df['rownum'] = df.groupby(['col_1','col_2']).cumcount()
df
```
