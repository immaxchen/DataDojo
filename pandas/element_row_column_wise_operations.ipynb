```python
import pandas as pd
```

```python
df = pd.DataFrame({'col_1':list('aabbbbbb'),
                   'col_2':list('aabbabab'),
                   'col_3':range(8),
                   'col_4':list(zip(range(8),range(1,9))),
                   'col_5':list(zip(list('abcdefgh'),list('12345678')))})
df
```

### element-wise operations on all cells

```python
df.applymap(str)
```

### element-wise operations on series

```python
df['col_4'] = df['col_4'].map(lambda x: sum(x))
df['col_5'] = df['col_5'].map(lambda x: ':'.join(x))
df
```

### row-wise combine strings from columns

```python
df['col_6'] = df['col_4'].map(str) + ':' + df['col_5'].map(str)
df
```

### row-wise operation by apply

```python
df[['col_3','col_4']].apply(lambda x: x.max()-x.min(), axis=1)
```

### column-wise operation by apply

```python
df[['col_3','col_4']].apply(lambda x: x.max()-x.min(), axis=0)
```
