```python
import numpy as np
import pandas as pd
```

```python
top_lv1 = range(3)
top_lv2 = ['col1', 'col2']
sub_lv1 = list('ab')
sub_lv2 = ['I', 'II']
```

```python
mii = pd.MultiIndex.from_product([top_lv1, sub_lv1, sub_lv2])
mic = pd.MultiIndex.from_product([top_lv2, sub_lv1])
df = pd.DataFrame(np.random.random([len(mii), len(mic)]), index=mii, columns=mic)
df
```

### direct slicing only works for top-level values

```python
df.loc[1:1, ['col2']] # OK
```

```python
df.loc[['b'], ['col2']] # not OK
```

```python
df.loc[1:1, ['b']] # not OK either
```

### use IndexSlice function to do it

```python
df.loc[pd.IndexSlice[:, ['b']], pd.IndexSlice[:, ['b']]]
```
